```python
import pandas as pd
```

```python
movie_merged = pd.read_csv('Movies_Merged.csv')
```

```python
movie_merged
```

```python
movie_merged['Release_Year'] = movie_merged['Release_Year'].astype('Int64')
```

```python
print(movie_merged['Release_Year'].head())
```

```python
movie_merged['profit'] = movie_merged['Revenue']- movie_merged['Budget'] # regression target
```

```python
movie_merged['hit'] = movie_merged['profit'].apply(lambda x: 1 if x > 0 else 0) # classification target
```

```python
numeric_col =['Budget', 'Runtime', 'Release_Year']
for col in numeric_col:
    movie_merged[col] = pd.to_numeric(movie_merged[col], errors='coerce'). fillna(0) # clean Numeric Columns
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
cat_col = ['Genre_Name','Language']
le_dict = {}
```

```python
for col in cat_col:
    movie_merged[col]= movie_merged[col]. fillna('unknown')
    le = LabelEncoder()
    movie_merged[col]=le.fit_transform(movie_merged[col])
    le_dict[col]= le    # Encode them using LabelEncoder
```

```python
features = numeric_col + cat_col
x = movie_merged[features]

y_class = movie_merged['hit'] # classification
y_reg = movie_merged['profit'] # Regression
```

```python
movie_merged.to_csv('Movies_Merged_clean.csv', index=False)
```

```python
import joblib
```

```python
joblib.dump(x, 'X_features.pkl')
joblib.dump(y_class, 'y_class.pkl')
joblib.dump(y_reg, 'y_reg.pkl')
```
