```python
import joblib
from xgboost import XGBClassifier, XGBRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, mean_absolute_error, r2_score
```

# Load saved features and targets

```python
x = joblib.load('X_features.pkl')
y_class = joblib.load('y_class.pkl')
y_reg = joblib.load('y_reg.pkl')
```

# Split data for Classification

```python
x_train, x_test, y_class_train, y_class_test = train_test_split(x,y_class, test_size=0.2, random_state=42) # Split data for classification
```

# Split data for Regression

```python
x_train, x_test, y_reg_train, y_reg_test = train_test_split(x,y_reg, test_size=0.2, random_state=42) # Split data for regression
```

# Initialize models

```python
xgb_class = XGBClassifier(
    n_estimators = 100,
    learning_rate = 0.1,
    max_depth = 5,
    random_state = 42
)
```

```python
xgb_reg = XGBRegressor(
    n_estimators = 100,
    learning_rate = 0.1,
    max_depth = 5,
    random_state = 42
)
```

# Train Models

```python
print('traning classification model')
xgb_class.fit(x_train,y_class_train)
```

```python
print('traning regression model')
xgb_reg.fit(x_train,y_reg_train)
```

# Evaluate models

## Classification

```python
y_pred_class = xgb_class.predict(x_test)
print("n\ classification report")
print('Accuracy', accuracy_score(y_class_test, y_pred_class))
print(classification_report(y_class_test, y_pred_class))
```

### Classification Check 

```python
t = xgb_class.score(x_train, y_class_train)
t1 = xgb_class.score(x_test, y_class_test)
print('train accuracy', t)
print('test accuracy', t1)
```

## Regression 

```python
y_pred_reg = xgb_reg.predict(x_test)
print("n\ regression evalution")
print('MAE', mean_absolute_error(y_reg_test, y_pred_reg))
print('R2 Score',r2_score(y_reg_test, y_pred_reg))
```

# Save Trained Models

```python
joblib.dump(xgb_class,'xgb_class_model.pkl')
joblib.dump(xgb_reg, 'xgb_reg_model.pkl')
```
